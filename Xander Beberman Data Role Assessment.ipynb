```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
donations = pd.read_csv('data/donations.csv')
```

```python
donations.head()
```

```python
donations.replace({'t':1, 'f':0}, inplace=True)
```

```python
donations.describe()
```

```python
# outputting count information
print('Column\t\t\tUnique values\tNA values')
for col in donations:
    print('{:21.21}\t{}\t\t{}'.format(donations[col].name, donations[col].nunique(), donations[col].isna().sum()))
```

```python
donations.hist(layout=(6,2), figsize=(8,24))
```

For small categorical variables

```python
fig = plt.figure(figsize=(12,6))
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)

ax1 = plt.subplot(grid[0,:])
ax2 = plt.subplot(grid[1,0])
ax3 = plt.subplot(grid[1,1:])

donations['donor_state'].value_counts().plot(kind='bar', ax=ax1, title='donor_state')
donations['dollar_amount'].value_counts().plot(kind='bar', ax=ax2, title='dollar_amount')
donations['payment_method'].value_counts().plot(kind='bar', ax=ax3, title='payment_method')
```

```python
donations.hist(column='donation_total', figsize=(12,4), log=True, bins=25)
```

```python
donations.hist(column='donation_total', range=(-10,25000), log=True, figsize=(12,4), bins=25)
```
